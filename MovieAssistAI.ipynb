### Implementing Intent Clarity and Intent Confirmation Layers


Let's start with the first part of the implementation - building the `intent clarity` and `intent confirmation` layers. As mentioned earlier, this layer helps in identifying the user requirements and passing it on to the product mapping layer. Here are the functions that we would be using for building these layers:

- `initialize_conversation()`

This initializes the variable conversation with the system message. Using prompt engineering and chain of thought reasoning, the function will enable the chatbot to keep asking questions until the user requirements have been captured in a dictionary. It also includes Few Shot Prompting(sample conversation between the user and assistant) to align the model about user and assistant responses at each step.

```python
# Import the libraries
import os, json, ast
import openai
from tenacity import retry, wait_random_exponential, stop_after_attempt
```

```python
def initialize_conversation():
    '''
    Returns a list [{"role": "system", "content": system_message}]
    '''

    delimiter = "####"

    example_user_dict = {'Duration': "high",
                        'title_year':"high",
                        'imdb_score': "low",
                        'aspect_ratio': "low",
                        'Genre': "Adventure"}

    example_user_req = {'Duration': "_",
                        'title_year': "_",
                        'imdb_score': "_",
                        'aspect_ratio': "_",
                        'Genre': "_"}

    system_message = f"""
    You are an intelligent movie recommendation expert and your goal is to find the best movie for a user.
    You need to ask relevant questions and understand the user profile by analysing the user's responses.
    You final objective is to fill the values for the different keys ('Duration','title_year','imdb_score','aspect_ratio','Genre') in the python dictionary and be confident of the values.
    These key value pairs define the user's profile.
    The python dictionary looks like this
    {{'Duration': 'values','title_year': 'values','imbd_score': 'values','aspect_ratio': 'values','Genre': 'values'}}
    The value for 'Genre' should be a text value extracted from the user's response.
    The values for all keys, except 'Genre', should be 'low', 'medium', or 'high' based on the importance of the corresponding keys, as stated by user.
    All the values in the example dictionary are only representative values.
    {delimiter}
    Here are some instructions around the values for the different keys. If you do not follow this, you'll be heavily penalised:
    - The values for all keys, except 'Genre', should strictly be either 'low', 'medium', or 'high' based on the importance of the corresponding keys, as stated by user.
    - The value for 'Genre' should be a text value extracted from the user's response.
    - 'Genre' value needs to be something from the list ['Adventure', 'Thriller', 'Sci-Fi', 'Romance', 'Animation','Fantacy', 'Action', 'Fantasy'].
     If the user says any other Genre than that, please mention that there are no movies in that Genre.
    - Do not randomly assign values to any of the keys.
    - The values need to be inferred from the user's response.
    {delimiter}

    To fill the dictionary, you need to have the following chain of thoughts:
    Follow the chain-of-thoughts below and only output the final updated python dictionary for the keys as described in {example_user_req}. \n
    {delimiter}
    Thought 1: Ask a question to understand the user's profile and requirements. \n
    If the duration for the movie is unclear. Ask followup questions to understand their needs.
    You are trying to fill the values of all the keys {{'Duration','title_year','imdb_score','aspect_ratio','Genre'}} in the python dictionary by understanding the user requirements.
    Identify the keys for which you can fill the values confidently using the understanding. \n
    Remember the instructions around the values for the different keys.
    If the necessary information has been extracted, only then proceed to the next step. \n
    Otherwise, rephrase the question to capture their profile clearly. \n

    {delimiter}
    Thought 2: Now, you are trying to fill the values for the rest of the keys which you couldn't in the previous step.
    Remember the instructions around the values for the different keys.
    Ask questions you might have for all the keys to strengthen your understanding of the user's profile.
    If yes, move to the next Thought. If no, ask question on the keys whose values you are unsure of. \n
    It is a good practice to ask question with a sound logic as opposed to directly citing the key you want to understand value for.
    {delimiter}

    {delimiter}
    Thought 3: Check if you have correctly updated the values for the different keys in the python dictionary.
    If you are not confident about any of the values, ask clarifying questions.
    {delimiter}

    {delimiter}
    Here is a sample conversation between the user and assistant:
    User: "Hi, I am a movie lover and like to watch movies."
    Assitant: "Great! As a movie lover, you likely need a good movie to watch. May I know your preferred duration for the movie in hours?"
    User: "Yes, movie duration should not be more than 2.5 hours"
    Assistant: "Thank you for the information. May I know which year movie you focus on? Are you interested to watch old movie or new one? Understanding the title year will help me tailor my recommendations accordingly. Let me know if my understanding is correct until now."
    User: "I would perfer movie with title year between 2009 to 2011."
    Assistant: "Thank you for providing that information. Do you have any requirement regarding imdb_score of the movie? Understanding your imdb_score will help in further classification of the movie."
    User: "Yes, any movie with imbd_score greater than 6 would work for me."
    Assistant: "Thank you for the information. Movie with atleast 6 imdb score would have good aspect_ratio. I have one more question, Do you have any requirement regarding aspect_ratio of the movie"
    User: "Yes, any movie with aspect_ratio greater than 2 would be good."
    Assistant:"Could you kindly let me know your favorite Genre? This will help me find options that fit within taste while meeting the specified requirements."
    User: "I would prefer Adventure genre"
    Assistant: "{example_user_dict}"
    {delimiter}

    Start with a short welcome message and encourage the user to share their requirements.
    """
    conversation = [{"role": "system", "content": system_message}]
    # conversation = system_message
    return conversation
```

- `get_chat_completions()`: This takes the ongoing conversation as the input and returns the response by the assistant. We'll use the Chat Completions function for performing LLM calls to OpenAI.

```python
# Define a Chat Completions API call
# Retry up to 6 times with exponential backoff, starting at 1 second and maxing out at 20 seconds delay
@retry(wait=wait_random_exponential(min=1, max=20), stop=stop_after_attempt(6))
def get_chat_completions(input, json_format = False):
    MODEL = 'gpt-3.5-turbo'

    system_message_json_output = """<<. Return output in JSON format to the key output.>>"""

    # If the output is required to be in JSON format
    if json_format == True:
        # Append the input prompt to include JSON response as specified by OpenAI
        input[0]['content'] += system_message_json_output

        # JSON return type specified
        chat_completion_json = openai.chat.completions.create(
            model = MODEL,
            messages = input,
            response_format = { "type": "json_object"},
            seed = 1234)

        output = json.loads(chat_completion_json.choices[0].message.content)

    # No JSON return type specified
    else:
        chat_completion = openai.chat.completions.create(
            model = MODEL,
            messages = input,
            seed = 2345)

        output = chat_completion.choices[0].message.content

    return output
```

### iterate_response() - Helper Function:
We've created a small helper test function to ensure the model's response is consistent.


```python
def iterate_llm_response(funct, debug_response, num = 10):
    """
    Calls a specified function repeatedly and prints the results.
    This function is designed to test the consistency of a response from a given function.
    It calls the function multiple times (default is 10) and prints out the iteration count,
    the function's response(s).
    Args:
        funct (function): The function to be tested. This function should accept a single argument
                          and return the response value(s).
        debug_response (dict): The input argument to be passed to 'funct' on each call.
        num (int, optional): The number of times 'funct' will be called. Defaults to 10.
    Returns:
        This function only returns the results to the console.
    """
    i = 0  # Initialize counter

    while i < num:  # Loop to call the function 'num' times

        response = funct(debug_response)  # Call the function with the debug response

        # Print the iteration number, result, and reason from the response
        print("Iteration: {0}".format(i))
        print(response)
        print('-' * 50)  # Print a separator line for readability
        i += 1  # Increment the counter

# Example usage: test the consistency of responses from 'intent_confirmation_layer'
# iterate_llm_response(get_chat_completions, messages)
```

### `moderation_check()`:
 This checks if the user's or the assistant's message is inappropriate. If any of these is inappropriate, you can add a break statement to end the conversation.

```python
# Define a function called moderation_check that takes user_input as a parameter.

def moderation_check(user_input):
    # Call the OpenAI API to perform moderation on the user's input.
    response = openai.moderations.create(input=user_input)

    # Extract the moderation result from the API response.
    moderation_output = response.results[0].flagged
    # Check if the input was flagged by the moderation system.
    if response.results[0].flagged == True:
        # If flagged, return "Flagged"
        return "Flagged"
    else:
        # If not flagged, return "Not Flagged"
        return "Not Flagged"
```

As mentioned earlier, you need to understand the user's profile, which essentially means that all the features: Duration, title_year, IMDb score, aspect ratio and Genre are captured or not. Let's look at the function that helps us verify that.

### `intent_confirmation_layer()`:

This function takes the assistant's response and evaluates if the chatbot has captured the user's profile clearly. Specifically, this checks if the following properties for the user has been captured or not
   - Duration
   - Title Year
   - IMDb Score
   - Aspect Ratio
   - Genre


```python
def intent_confirmation_layer(response_assistant):

    delimiter = "####"

    allowed_values = {'low','medium','high'}

    prompt = f"""
    You are a senior evaluator who has an eye for detail.The input text will contain a user requirement captured through 5 keys.
    You are provided an input. You need to evaluate if the input text has the following keys:
    {{
    'Duration': 'values',
    'title_year':'values',
    'imdb_score':'values',
    'aspect_ratio':'values',
    'Genre':'values'
    }}
    The values for the keys should only be from the allowed values: {allowed_values}.
    The 'Genre' key can take only value from the list ['Adventure', 'Thriller', 'Sci-Fi', 'Romance', 'Animation','Fantacy', 'Action', 'Fantasy'].
    Next you need to evaluate if the keys have the the values filled correctly.
    Only output a one-word string in JSON format at the key 'result' - Yes/No.
    Thought 1 - Output a string 'Yes' if the values are correctly filled for all keys, otherwise output 'No'.
    Thought 2 - If the answer is No, mention the reason in the key 'reason'.
    THought 3 - Think carefully before the answering.
    """

    messages=[{"role": "system", "content":prompt },
              {"role": "user", "content":f"""Here is the input: {response_assistant}""" }]

    response = openai.chat.completions.create(
                                    model="gpt-3.5-turbo",
                                    messages = messages,
                                    response_format={ "type": "json_object" },
                                    seed = 1234
                                    # n = 5
                                    )

    json_output = json.loads(response.choices[0].message.content)

    return json_output
```

### `dictionary_present()`:

This function checks if the final understanding of user's profile is returned by the chatbot is a Python dictionary or not. This is important as it'll be used later on for finding the right movies using dictionary matching.

```python
def dictionary_present(response):
    delimiter = "####"

    user_req = {'Duration': 'high',
                'title_year': 'high',
                'imdb_score': 'medium',
                'aspect_ratio': 'high',
                'Genre': 'Adventure'}

    prompt = f"""You are a python expert. You are provided an input.
            You have to check if there is a python dictionary present in the string.
            It will have the following format {user_req}.
            Your task is to just extract the relevant values from the input and return only the python dictionary in JSON format.
            The output should match the format as {user_req}.

            {delimiter}
            Make sure that the value of Genre is also present in the user input. ###
            The output should contain the exact keys and values as present in the input.
            Ensure the keys and values are in the given format:
            {{
            'Duration': 'low/medium/high ',
            'title_year':'low/medium/high',
            'imdb_score':'low/medium/high',
            'aspect_ratio':'low/medium/high',
            'Genre':'Adventure/Thriller/Sci-Fi/Romance/Animation/Fantacy/Action/Fantasy'
            }}
            Here are some sample input output pairs for better understanding:
            {delimiter}
            input 1: - Duration: low - title_year: high - imdb_score: low - Multitasking: high - aspect_ratio: medium - Genre: Adventure
            output 1: {{'Duration': 'low', 'title_year': 'high', 'imdb_score': 'low', 'Multitasking': 'high', 'aspect_ratio': 'medium', 'Genre': 'Adventure'}}

            input 2: {{'Duration':     'low', 'title_year':     'low', 'imdb_score':    'medium', 'Multitasking': 'medium', 'aspect_ratio': 'low', 'Genre': 'Thriller'}}
            output 2: {{'Duration': 'low', 'title_year': 'low', 'imdb_score': 'medium', 'Multitasking': 'medium', 'aspect_ratio': 'low', 'Genre': 'Thriller'}}

            input 3: Here is your user profile 'Duration': 'high','title_year': 'high','imdb_score': 'medium','Multitasking': 'high','aspect_ratio': 'high','Genre': 'Romance'
            output 3: {{'Duration': 'high','title_year': 'high','imdb_score': 'medium','Multitasking': 'high','aspect_ratio': 'high','Genre': 'Romance'}}
            {delimiter}
            """
    messages = [{"role": "system", "content":prompt },
                {"role": "user", "content":f"""Here is the user input: {response}""" }]

    confirmation = get_chat_completions(messages, json_format = True)

    return confirmation
```

### `product_map_layer()`:

This function is responsible for extracting key features and criteria from movie descriptions. Here's a breakdown of how it works:

-  Use a prompt that assign it the role of a movie Specifications Classifier, whose objective is to extract key features and classify them based on movie descriptions.

- Provide step-by-step instructions for extracting movie features from description.

- Assign specific rules for each feature (e.g., Duration, title_year, IMDb score, aspect ratio) and associate them with the appropriate classification value (Low, Medium, or High).

- Includes Few Shot Prompting (sample conversation between the user and assistant) to demonstrate the expected result of the feature extraction and classification process.

```python
def product_map_layer(movie_description):
    delimiter = "#####"

    movie_spec = {
        "Duration":"(Duration of the movie)",
        "title_year":"(title_year)",
        "imdb_Score":"(imdb_score)",
        "aspect_ratio":"(aspect_ratio)"
    }

    values = {'low','medium','high'}

    prompt=f"""
    You are a Movie Specifications Classifier whose job is to extract the key features of movie and classify them as per their requirements.
    To analyze each movie, perform the following steps:
    Step 1: Extract the movie's primary features from the description {movie_description}
    Step 2: Store the extracted features in {movie_spec} \
    Step 3: Classify each of the items in {movie_spec} into {values} based on the following rules: \
    {delimiter}
    Duration:
    - low: <<< if movie duration is between 1 to 2.5 >>> , \n
    - medium: <<< if movie duration is greater than 2.5 and less than to 3 >>> , \n
    - high: <<< if movie duration is greater than 3  >>> , \n

    title_year:
    - low: <<< if movie title_year is between 2006 and 2009 >>> , \n
    - medium: <<< if movie title_year is between 2010 and 2013 >>> , \n
    - high: <<< if movie title_year is between 2014 and 2016 >>> \n

    imdb_score:
    - low: <<< if movie imdb_score is between 6.1 and 7 >>> , \n
    - medium: <<< if movie imdb_score is between 7.1 and 8 >>> , \n
    - high: <<< if movie imdb_score is between 8.1 and 10 >>> \n

    aspect_ratio:
    - low: <<< if movie aspect_ratio is between 0 and 1 >>> , \n
    - medium: <<< if movie aspect_ratio is between 1.1 and 2 >>> , \n
    - high: <<< if movie aspect_ratio is greater than 2 >>> \n

    {delimiter}

    {delimiter}
    Here is input output pair for few-shot learning:
    input 1: "The movie name is Avatar. Its an aventurours movie. It is made of 237000000 dollars budget. The main words that can descibe the movie are Avatar, Future, Marine, Native, Paraplegic. It's in English language. It's released year is 2009. It's duration is 2.97 hours. It has IMDb rating of 7.9. It has aspect_ratio of 1.78."
    output 1: {{'Duration': 'medium','title_year':'low','imdb_score':'medium','aspect_ratio':'medium'}}

    {delimiter}
    ### Strictly don't keep any other text in the values of the JSON dictionary other than low or medium or high ###
    """
    input = f"""Follow the above instructions step-by-step and output the dictionary in JSON format {movie_spec} for the following movie {movie_description}."""
    #see that we are using the Completion endpoint and not the Chatcompletion endpoint
    messages=[{"role": "system", "content":prompt },{"role": "user","content":input}]

    response = get_chat_completions(messages, json_format = True)

    return response
```

### `compare_movie_with_user()`:

This function compares the user's profile with the different movie and come back with the top  recommendations. It will perform the following steps:
    - It will take the user requirements dictionary as input
    - Filter the movie based on their Genre, keeping only the ones with the user's Genre.
    - Calculate a score for each movie based on how well it matches the user's requirements.
    - Sort the movie based on their scores in descending order.
    - Return the top 3 movies as a JSON-formatted string.

```python
import pandas as pd  # Importing the pandas library for data manipulation
import json
def compare_movie_with_user(response_dict_n):
    movie_df = pd.read_csv('updated_movie.csv')

    user_requirements = response_dict_n

    # Extracting user requirements from the input string (assuming it's a dictionary)
    # Since the function parameter already seems to be a string, we'll use it directly instead of extracting from a dictionary

    # Extracting the budget value from user_requirements and converting it to an integer
    Genre= str(user_requirements.get('Genre'))
    # budget = int(user_requirements.get('Budget', '0').replace(',', '').split()[0])

    # budget
    # # Creating a copy of the DataFrame
    filtered_movies = movie_df.copy()

    filtered_movies = filtered_movies[filtered_movies['Genre'] == Genre].copy()

    # # # Mapping string values 'low', 'medium', 'high' to numerical scores 0, 1, 2
    mappings = {'low': 0, 'medium': 1, 'high': 2}



    # # # Creating a new column 'Score' in the filtered DataFrame and initializing it to 0
    filtered_movies['Score'] = 0

    # # # Iterating over each movie in the filtered DataFrame to calculate scores based on user requirements
    for index, row in filtered_movies.iterrows():
        user_product_match_str = row['movie_feature']
        movie_values = user_product_match_str
        movie_values = dictionary_present(user_product_match_str)
        score = 0

    #     # Comparing user requirements with movie features and updating scores
        for key, user_value in user_requirements.items():
            # if key.lower() == 'budget':
            if key == 'Genre':
                continue  # Skipping budget comparison
            movie_value = movie_values.get(key, None)
            # print(key, movie_value)
            movie_mapping = mappings.get(movie_value, -1)
            user_mapping = mappings.get(user_value, -1)
            if movie_mapping >= user_mapping:
                score += 1  # Incrementing score if movie value meets or exceeds user value

        filtered_movies.loc[index, 'Score'] = score  # Updating the 'Score' column in the DataFrame

    # Sorting movies  by score in descending order and selecting the top 3 products
    top_movies = filtered_movies.drop('movie_feature', axis=1)
    top_movies = top_movies.sort_values('Score', ascending=False).head(3)
    top_movies_json = top_movies.to_json(orient='records')  # Converting the top movies DataFrame to JSON format

    # top_movies
    return top_movies_json
```

Finally, we come to the product recommendation layer. It takes the output from the `compare_movie_with_user` function in the previous layer and provides the recommendations to the user. It has the following steps.

```python
def initialize_conv_reco(products):
    system_message = f"""
    You are an intelligent movie recommendation expert and you are tasked with the objective to \
    solve the user queries about any product from the catalogue in the user message \
    You should keep the user profile in mind while answering the questions.\

    Start with a brief summary of each movie in the following format, in decreasing order of Duration of Movies:
    1. <Movie Name> : <Major specifications of the Movie>, <Duration in Hours>
    2. <Movie Name> : <Major specifications of the Movie>, <Duration in Hours>

    """
    user_message = f""" These are the user's products: {products}"""
    conversation = [{"role": "system", "content": system_message },
                    {"role":"user","content":user_message}]
    # conversation_final = conversation[0]['content']
    return conversation
```
